```python
td.info()
```

```python
#1,000 =1k
#1,000,000=1m
#1,000,000,000=1b
def preprocess_object_value(value):
    if isinstance(value, str):
        value = value.lower().strip()
        if value.endswith('k'):
            return float(value[:-1]) * 1000
        elif value.endswith('m'):
            return float(value[:-1]) * 1000000
        elif value.endswith('b'):
            return float(value[:-1])* 1000000000
    return value
```

```python
j=0
for i in td['거래량C']:
  a=preprocess_object_value(i)
  a=pd.to_numeric(a)/100
  td['거래량C'][j]=a
  j+=1
td['거래량C']=pd.to_numeric(td['거래량C'])
```
