```python
import pandas as pd
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
senti_analyzer=SentimentIntensityAnalyzer()
```

# 감성분석 함수 생성

```python
def get_sentiment(sentence):
    senti_score=senti_analyzer.polarity_scores(sentence)
    return(senti_score['compound'])
```

# 감성분석 수행

```python
cnn=pd.read_csv('CNN_기사_final.csv',encoding='cp949')
cnn['score']=cnn['title'].apply(get_sentiment)

a=cnn.groupby('date')['score'].mean()
a.index=pd.to_datetime(a.index)
```

# 신한 데이터와 합치기

```python
sh=pd.read_csv('신한지주_pre.csv')
sh['날짜']=pd.to_datetime(sh['날짜'])
sh_merge=pd.merge(sh,a,left_on='날짜',right_on=a.index,how='outer')
```
